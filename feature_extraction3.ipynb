```python
import numpy as np
import librosa
import librosa.display
import librosa.core
import IPython.display as ipd
import matplotlib.pyplot as plt
import soundfile as sf
import os 
import pandas as pd
import os
import glob
```

```python
# Define the directory path
directory = 'C:\\Users\\aaliy\\OneDrive\\Desktop\\Speech Processing Lab IIITH\\ml-stuttering-events-dataset\\Type_sep28k'
```

```python
# Define the types of folders you want to iterate over
folder_types = ['Block', 'DifficultToUnderstand', 'Interjection','Music','NaturalPause','NoSpeech','NoStutteredWords','PoorQualityAudio','Prolongation','SoundRep','Unsure','WordRep']
```

```python
# Iterate over the folder types
for folder_type in folder_types:
    # Get the path of folders for the current type
    folder_path = os.path.join(directory, folder_type)

    # Check if the folder exists
    if os.path.exists(folder_path) and os.path.isdir(folder_path):
        # Iterate over the files in the current folder
        for file_path in glob.glob(os.path.join(folder_path, '*')):
            # Process the file as needed
            print(file_path)
            audio, sample_rate = librosa.load(file_path)
            mfcc_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13)
            mfcc_mean = np.mean(mfcc_features, axis=1)
            print(mfcc_mean)
```
