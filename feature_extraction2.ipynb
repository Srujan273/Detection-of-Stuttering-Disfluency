```python
def visualize_mfcc(mfccs,sr):
    plt.figure(figsize=(25, 10))
    librosa.display.specshow(mfccs, 
                         x_axis="time", 
                         sr=sr)
    plt.colorbar(format="%+2.f")
    plt.show()
```
