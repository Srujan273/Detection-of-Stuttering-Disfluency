```python
import re

def extract_parts_from_path(file_path):
    # Define a regular expression pattern to extract the individual parts.
    pattern = r"\\([^\\]+)\\([^\\]+)_(\d+)_(\d+)\.wav$"
    
    # Use re.search to find the pattern in the file_path.
    match = re.search(pattern, file_path)
    
    if match:
        # Extract the individual parts from the regex match object.
        part1 = match.group(2)  # WomenWhoStutter
        part2 = match.group(1)  # Block
        part3 = match.group(3)  # 105
        part4 = match.group(4)  # 26
        return part1, part2, f"{part3}_{part4}"
    else:
        # If no match is found, return None or raise an error, depending on your use case.
        return None, None
```
